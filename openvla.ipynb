```python
# Install minimal dependencies (`torch`, `transformers`, `timm`, `tokenizers`, ...)
# > pip install -r https://raw.githubusercontent.com/openvla/openvla/main/requirements-min.txt
from transformers import AutoModelForVision2Seq, AutoProcessor
from PIL import Image

import torch
```

```python
# Load Processor & VLA
processor = AutoProcessor.from_pretrained("openvla/openvla-7b", trust_remote_code=True)
vla = AutoModelForVision2Seq.from_pretrained(
    "openvla/openvla-7b",
    # attn_implementation="flash_attention_2",  # [Optional] Requires `flash_attn`
    torch_dtype=torch.bfloat16,
    low_cpu_mem_usage=True,
    trust_remote_code=True
).to("cuda:0")

```

```python

image: Image.Image = Image.open("testimg.png")   # You should replace this with your image acquisition code
prompt = "In: What action should the robot take to pick up the red block and place it in the blue bowl?\nOut:"

# Predict Action (7-DoF; un-normalize for BridgeData V2)
inputs = processor(prompt, image).to("cuda:0", dtype=torch.bfloat16)
action = vla.predict_action(**inputs, unnorm_key="bridge_orig", do_sample=False)
```

```python
import base64
import io
```

```python
# Inference function
def predict_action(base64_image: str, instruction: str):
    image_bytes = base64.b64decode(base64_image)
    image = Image.open(io.BytesIO(image_bytes)).convert("RGB")

    prompt = f"In: What action should the robot take to {instruction}?\nOut:"
    inputs = processor(prompt, image).to("cuda:0", dtype=torch.float16)
    action = vla.predict_action(**inputs, unnorm_key="bridge_orig", do_sample=False)
    return action.tolist()
```

```python
from fastapi import FastAPI, Request
from pyngrok import ngrok
import uvicorn

app = FastAPI()

@app.post("/predict")
async def predict(req: Request):
    data = await req.json()
    base64_image = data["image"]
    instruction = data["instruction"]
    result = predict_action(base64_image, instruction)
    return {"action": result}

# Open public tunnel
public_url = ngrok.connect(8000)
print("Public URL:", public_url)

uvicorn.run(app, host="0.0.0.0", port=8000)
```
